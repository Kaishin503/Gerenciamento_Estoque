```python
import pyodbc
conn = pyodbc.connect(
    "DRIVER={ODBC Driver 17 for SQL Server};"
    "SERVER=DESKTOP-SIQD2J5;"
    "DATABASE=CRUD_DATABASE;"
    "Trusted_Connection=yes;")
cursor = conn.cursor()

```

```python
print("Welcome! Below Are The Numbers For Each Functionality")
print("----------------------------------------------------------------")
print("1 - Register Product | 2 - Register Category | 3 - List Products ")
print("4 - Search Product | 5 - Delete Product | 6 - Update Product  ")
print("7 - Inventory Log (Unavailable)\n")
import string

while True:
    while True:
        try:
            option = int(input("Type The Function Number:"))
            if option not in [1,2,3,4,5,6,0]:
                print("ERROR: Choose A Valid Option!\n")
                continue
            break
        except ValueError:
            print("ERROR: Type A Valid Number!\n")
            continue
    if option == 1:
        from CRUDMAINFUNCTIONS import create_product
        allowed_characters = string.ascii_lowercase + string.whitespace
        while True:
            product_name = input("Name:")
            if product_name == "":
                print("Operation Canceled.\n")
                break
            if product_name not in allowed_characters:
                print("ERROR: Name Cannot Contain Any Special Characters Or Numbers!\n")
                continue
            product_name = product_name.lower()
            break
        if not product_name:
            continue
        while True:
            category = input("Category:")
            if category == "":
                print("Operation Canceled.\n")
                break
            if category not in allowed_characters:
                print("ERROR: Category Name Cannot Contain Any Special Characters Or Numbers!\n")
                continue
            category = category.lower()
            break
        if not category:
            continue
        while True:
            try:
                price = input("Price:")
                if not price:
                    print("Operation Canceled.\n")
                    break
                price = float(price)
                break
            except ValueError:
                print("ERROR: Invalid Character!\n")
                continue
        if not price:
            continue
        while True:
            try:
                quantity = input("Initial Quantity:")
                if not quantity:
                    print("Operation Canceled.\n")
                    break
                quantity = int(quantity)
                break
            except ValueError:
                print("ERROR: Invalid Character!\n")
                continue
        if not quantity:
            continue
            
    elif option == 2:
        from CRUDMAINFUNCTIONS import create_category
        while True:
            category_name = input("Name Of The Category:")
            if category_name == "":
                print("Operation Canceled.\n")
                break
            if category_name.isalpha() == False:
                if " " in category_name:
                    pass
                else:
                    print("ERROR: Name Cannot Contain Numbers Or Invalid Characters!\n")
                    continue
            category_name = category_name.lower()
            try:
                category_creation = create_category(cursor,category_name)
                print(f"{category_creation}\n")
                break
            except ValueError as message:
                print(f"{message}")

    elif option == 3:
        from CRUDMAINFUNCTIONS import list_products
        products = list_products(cursor)
        for item in products:
            id,name,category_name,categoryFK,price,stock = item
            print("ID:{}\nName: {}\nCategory: {}\nCategoryID: {}\nPrice: ${}\nStock: {} Units\n".format(id,name,category_name,categoryFK,price,stock))
    
    elif option == 4:
        from CRUDMAINFUNCTIONS import search_product
        id_or_name = input("Name Or ID Of The Product:")
        if id_or_name == "":
            print("Operation Canceled.\n")
            continue
        id_or_name = id_or_name.lower()
        try:
            product = search_product(cursor,id_or_name)
        except ValueError as message:
            print(f"{message}\n")
        id,name,category_name,categoryFK,price,stock = product
        print("ID:{}\nName: {}\nCategory: {}\nCategoryID: {}\nPrice: ${}\nStock: {} Units\n".format(id,name,category_name,categoryFK,price,stock))
    
    elif option == 5:
       from CRUDMAINFUNCTIONS import update_product
       update_product(cursor)
    elif option == 6:
        from CRUDMAINFUNCTIONS import delete_product
        delete_product(cursor)
        conn.commit()
    elif option == 0:
        print("Leaving...")
        conn.close()
        break
    
```
